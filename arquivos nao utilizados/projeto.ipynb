```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carregar os arquivos CSV
df_2020 = pd.read_csv('dados_2020.csv')
df_2021 = pd.read_csv('dados_2021.csv')
df_2022 = pd.read_csv('dados_2022.csv')
```

```python
# Filtrar as colunas relacionadas ao ponto de virada
# Converter "Sim" para 1 e "Não" para 0
df_2020['PONTO_VIRADA_2020'] = df_2020['PONTO_VIRADA_2020'].map({'Sim': 1, 'Não': 0})
df_2021['PONTO_VIRADA_2021'] = df_2021['PONTO_VIRADA_2021'].map({'Sim': 1, 'Não': 0})
df_2022['PONTO_VIRADA_2022'] = df_2022['PONTO_VIRADA_2022'].map({'Sim': 1, 'Não': 0})

# Calcular a porcentagem de "Sim" e "Não" para cada ano
percent_virada_2020 = df_2020['PONTO_VIRADA_2020'].value_counts(normalize=True) * 100
percent_virada_2021 = df_2021['PONTO_VIRADA_2021'].value_counts(normalize=True) * 100
percent_virada_2022 = df_2022['PONTO_VIRADA_2022'].value_counts(normalize=True) * 100

# Exibir as porcentagens
print("2020 - Porcentagens de 'Sim' e 'Não':")
print(percent_virada_2020)

print("\n2021 - Porcentagens de 'Sim' e 'Não':")
print(percent_virada_2021)

print("\n2022 - Porcentagens de 'Sim' e 'Não':")
print(percent_virada_2022)
```

```python
# Função para criar o gráfico de pizza
def plot_pie_chart(counts, year):
    labels = ['Sim', 'Não']
    colors = ['#66b3ff','#ff9999']
    plt.figure(figsize=(8, 8))
    plt.pie(counts, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors)
    plt.title(f'Ponto de Virada - {year}')
    plt.show()
```

```python
# Calcular as contagens de "Sim" e "Não" para cada ano
counts_2020 = df_2020['PONTO_VIRADA_2020'].value_counts()
counts_2021 = df_2021['PONTO_VIRADA_2021'].value_counts()
counts_2022 = df_2022['PONTO_VIRADA_2022'].value_counts()

# Gerar os gráficos de pizza para 2020, 2021 e 2022
plot_pie_chart(counts_2020, '2020')
plot_pie_chart(counts_2021, '2021')
plot_pie_chart(counts_2022, '2022')
```
