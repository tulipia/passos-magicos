```python
import pandas as pd

# Carregar o dataset (substitua 'seu_arquivo.csv' pelo nome do arquivo correto)
df = pd.read_csv('PEDE_PASSOS_DATASET_FIAP.csv', delimiter=';')

# Filtrar colunas para cada ano
df_2020 = df.filter(regex='2020|NOME', axis=1)
df_2021 = df.filter(regex='2021|NOME', axis=1)
df_2022 = df.filter(regex='2022|NOME', axis=1)

# Exibir as primeiras linhas para verificar
print("Dados de 2020:")
print(df_2020.head())

print("\nDados de 2021:")
print(df_2021.head())

print("\nDados de 2022:")
print(df_2022.head())
```

```python
def cleaning_dataset(df):
  _df = df.dropna(subset=df.columns.difference(['NOME']), how='all') # executa o dropna para todas as colunas sem visualizar a coluna NOME
  _df = _df[~_df.isna().all(axis=1)] # remove linhas com apenas NaN, se tiver algum dado na linha não remove
  return _df
```

```python
df_2020 = cleaning_dataset(df_2020)
df_2021 = cleaning_dataset(df_2021)
df_2022 = cleaning_dataset(df_2022)
```

```python
# Salvar os novos datasets em CSVs separados
df_2020.to_csv('dados_2020.csv', index=False)
df_2021.to_csv('dados_2021.csv', index=False)
df_2022.to_csv('dados_2022.csv', index=False)
```
